2. Imputasi

```python
print(df['Perceptions of corruption'].dropna().describe())
print(df['Generosity'].dropna().describe())
print(df['Log GDP per capita'].dropna().describe())
print(df['Social support'].dropna().describe())
print(df['Healthy life expectancy at birth'].dropna().describe())
print(df['Freedom to make life choices'].dropna().describe())
print(df['Positive affect'].dropna().describe())
print(df['Negative affect'].dropna().describe())
```

diatas adalah persentase nilai yang hilang dan selanjunya saya akan melakukan imputasi semua data nilainnya.

```python
df['Generosity'] = df['Generosity'].fillna(df['Generosity'].dropna().mean())
df['Social support'] = df['Social support'].fillna(df['Social support'].dropna().mean())
df['Log GDP per capita'] = df['Log GDP per capita'].fillna(df['Log GDP per capita'].dropna().mean())
df['Healthy life expectancy at birth'] = df['Healthy life expectancy at birth'].fillna(df['Healthy life expectancy at birth'].dropna().mean())
df['Perceptions of corruption'] = df['Perceptions of corruption'].fillna(df['Perceptions of corruption'].dropna().mean())
df['Positive affect'] = df['Positive affect'].fillna(df['Positive affect'].dropna().mean())
df['Freedom to make life choices'] = df['Freedom to make life choices'].fillna(df['Freedom to make life choices'].dropna().mean())
df['Negative affect'] = df['Negative affect'].fillna(df['Negative affect'].dropna().mean())
```

Diatas adalah imputasi nilai, disini saya langsung imputasi semua nilai.

```python
pd.DataFrame(df.isna().sum() / len(df) * 100, columns=['Null Ratio %'])
```

Diatas adalah hasil dari imputasi data nilai.

3. Outliers

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_df)
```

Dari hasil diatas saya akan melakukan imputasi data.


Imputasi Outliers

```python
columns_to_impute = ["Social support", "Healthy life expectancy at birth","Generosity","Perceptions of corruption","Negative affect"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR


    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

## Visualisasi Data yang Relevan dengan Project Understanding

### Menghitung Tren Kabahagian Global dari tahun ke tahun

```python
life_trend = df.groupby("year")["Life Ladder"].mean()

plt.figure(figsize=(12, 6))
sns.lineplot(x=life_trend.index, y=life_trend.values, marker="o", linestyle="-", color="skyblue")
plt.title("Tren Kebahagiaan Global dari Tahun ke Tahun")
plt.xlabel("Tahun")
plt.ylabel("Rata-rata Life Ladder")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

Tingkat kebahagian dunia itu mengalami naik turun dari tahun ke tahun, kebahagian mencapai puncak turunya itu terjadi di sekitar tahun 2005, namun sejak tahun tahun 2007 tingkat kebahagian mulai meningkat lagi.

Tindakkan yang bisa dilakukan ialah menganalisis faktor faktor apa saja yang menyebabkan terjadinya penurunan pada tahun tahun tertentu.

### Hubungan GDP Per Kapita dan kebahagiaan

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Log GDP per capita'], y=df['Life Ladder'], alpha=0.6, color='blue')

plt.title("Hubungan GDP per Kapita dan Kebahagiaan")
plt.xlabel("Log GDP per Kapita")
plt.ylabel("Life Ladder (Kebahagiaan)")
plt.grid(True)
plt.show()
```

Hubungan GDP per Kapita dan Kebahagiaan:

Negara dengan GDP per kapitanya tinggi itu lebih bahagia, tetapi beberapa negara dengan ekonomi yang kuat itu masih memiliki kebahagiaan rendah karena faktor sosial atau politik.

Tindakan yang bisa di ambil itu dengan melakukan peningkatan kesejahteraan sosial pada negara dengan GDP terendah yang dimana peningkatan ini bisa dilakukan tanpa harus sepenuhnya bergantung ke ekonomi dan juga bisa melakukan analisis tindak lanjut terkait faktor sosial atau politik.

### Hubungan antara Persepsi korupsi dengan kebahagiaan

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["Perceptions of corruption"], y=df["Life Ladder"], alpha=0.6, color="red")

plt.title("Hubungan Persepsi Korupsi dan Kebahagiaan")
plt.xlabel("Persepsi Korupsi (Semakin Tinggi = Lebih Korup)")
plt.ylabel("Life Ladder (Tingkat Kebahagiaan)")
plt.grid(True)
plt.show()
```

Hubungan persepsi korupsi dan kebahagian

Negara dengan tingkat korupsi yang tinggi memiliiki indeks kebahagian yang rendah

Tindakan yang bisa dilakukan ialah menerapkan transparasi politik agar masyarakat bisa lebih percaya terhadap pemerintah.

### Perbandingan Kebebasan hidup negara tertinggi dan terendah

```python
top_countries = df.groupby("Country name")["Freedom to make life choices"].mean().nlargest(10)
bottom_countries = df.groupby("Country name")["Freedom to make life choices"].mean().nsmallest(10)
freedom_df = pd.concat([top_countries, bottom_countries])

plt.figure(figsize=(12, 6))
sns.barplot(x=freedom_df.values, y=freedom_df.index, palette="viridis")
plt.title("Perbandingan Kebebasan Hidup (10 Negara Tertinggi & Terendah)")
plt.xlabel("Rata-rata Kebebasan Hidup")
plt.ylabel("Negara")
plt.show()
```

Perbandingan kebebasan

Negara dengan kebebasan tertinggi memliki tingkat kebahagian yang lebih tinggi dibandingkan negara dengan tingkat kebebasan yang rendah

Tindakan yang bisa dilakukan ialah dengan menerapapkan kebebasan ekspresi dan mendukung penuh atas hak asasi manusia
